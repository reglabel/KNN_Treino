```python
#manipulacao de dados
import pandas as pd
import numpy as np
```

```python
test = pd.read_csv("datasets/test.csv")
train = pd.read_csv("datasets/train.csv")
train.head()
```

```python
train.info()
```

```python
train.describe()
```

```python
#separando a coluna de labels(rotulos) para o Y
Y = train['label']
print(Y)
```

```python
#X fica com só com os valores de pixels
X = train.drop(labels=['label'], axis=1)
print(X.head())
```

```python
x = X.values.reshape(42000, 784) #matriz de 42000 linhas com 784 colunas
x = x.astype("float32") #convertendo os valores pro tipo float
x /= 255 #dividindo pelo valor máximo(255-rgb) para manter na mesmo proporção em uma escala de 0 a 1

print(x[117])
```

```python
from tensorflow import keras
```

```python
num_classes = 10 #numero de classes
y = keras.utils.to_categorical(Y, num_classes) #criando as categorias

print(y[117])
```

```python
#Separando o teste dos dados já tratados e utilizando KNN
```

```python
from sklearn.model_selection import train_test_split 
X_train,X_test,y_train,y_test = train_test_split(x,y, test_size=0.3)
```

```python
#Interessante...ao passor por esse modificador de escala o modelo teve 0.92 de acurácia. Ao ignorar essa parte, ele teve 0.96

#from sklearn.preprocessing import StandardScaler  
#escala = StandardScaler()  
#escala.fit(X_train)

#X_train = escala.transform(X_train)  
#X_test = escala.transform(X_test)
```

```python
#Importando o modelo KNN
from sklearn.neighbors import KNeighborsClassifier  

# Definindo o valor de visinhos
classifier = KNeighborsClassifier(n_neighbors=5)  

#Treinando o modelo, com dados de treinamento
classifier.fit(X_train, y_train) 
```

```python
y_pred = classifier.predict(X_test)
```

```python
# Importando métricas para validação do modelo
from sklearn.metrics import classification_report, accuracy_score

# Imprimindo o relatório de classificação
print("Relatório de classificação: \n", classification_report(y_test, y_pred))  


# Imprimindo o quão acurado foi o modelo
print('Acurácia do modelo: ' , accuracy_score(y_test, y_pred))
```
